# **Final Project Proposal Guidelines**

## **Objective**
The purpose of this proposal is to outline your final project, ensuring it is well-defined, feasible, and aligned with the course objectives. Your project must:  
1. **Be related to finance or corporate finance.**  
2. **Incorporate an ML or AI model** covered in class.  

Your proposal should clearly define the problem, methodology, and expected outcomes.  

---

## **Proposal Requirements**
Your proposal should be **2-3 pages** (excluding references) and must include the following sections:  

### **1. Project Title**  
Provide a clear and concise title that reflects your project topic.  

### **2. Introduction & Problem Statement**  
- Explain the financial or corporate finance problem you aim to solve.  
- Justify why this problem is important and relevant.  
- Provide background information, including any existing solutions or gaps.  

### **3. Data Description**  
- Describe the dataset you will use (source, size, and key variables).  
- If you are collecting or synthesizing data, explain your approach.  
- Discuss any necessary preprocessing steps (e.g., handling missing values, feature selection).  

### **4. Methodology**  
- Specify the **ML or AI model(s)** you plan to use and why they are appropriate.  
- Outline the training and evaluation process.  
- Mention any baseline models for comparison.  

### **5. Expected Outcomes & Evaluation Metrics**  
- Define what success looks like (e.g., improved prediction accuracy, financial insights).  
- Specify evaluation metrics (e.g., RMSE, accuracy, precision-recall, Sharpe ratio).  

### **6. Challenges & Considerations**  
- Discuss potential obstacles (e.g., data limitations, computational complexity).  
- Suggest possible solutions or alternative approaches.  

### **7. Timeline & Next Steps**  
- Provide a high-level timeline of key milestones (e.g., data collection, model training, evaluation).  
- Outline immediate next steps.  

### **8. References (if applicable)**  
- List any papers, datasets, or resources referenced in your proposal.  

---

## **Submission Details**
- **Format:** PDF or Word document  
- **Length:** 2-3 pages (excluding references)  
- **Deadline:** 3/3/2025 
- **Submission Method:** Upload to D2L

## **Grading Criteria**
Your proposal will be evaluated based on the following:  
- **Clarity & Structure (20%)** – Well-organized and clearly written.  
- **Relevance (20%)** – Strong connection to finance/corporate finance.  
- **Methodology (30%)** – Well-justified ML/AI approach.  
- **Feasibility (20%)** – Realistic given data and timeline.  
- **Potential Impact (10%)** – Meaningful financial insights or applications.
