```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error


# Set random seed for reproducibility
np.random.seed(42)

# Create a small dataset with 3 features
n_sample = 15
house_size = np.random.randint(1000, 3000, n_sample)  # House size in sq ft
num_bedrooms = np.random.randint(2, 5, n_sample)  # Number of bedrooms
year_built = np.random.randint(1980, 2022, n_sample)  # Year built
house_price = house_size * 100 + num_bedrooms * 5000 + (2022 - year_built) * (-200) + np.random.normal(0, 10000, n_sample)

# Create DataFrame
df = pd.DataFrame({"House Size": house_size, "Num Bedrooms": num_bedrooms, "Year Built": year_built, "House Price": house_price})
print("Original Dataset:")
print(df)

# Initial guess: Mean of house prices
initial_guess = np.mean(df["House Price"])
df["Prediction"] = initial_guess
print(f"\nInitial Guess (Mean Price): {initial_guess}")

# Store MSE values
mse_values = []
mse_values.append(mean_squared_error(df["House Price"], df["Prediction"]))

# Learning rate
learning_rate = 0.05
num_trees = 20


# Store models
trees = []

for i in range(num_trees):
    # Compute residuals
    df["Residual"] = df["House Price"] - df["Prediction"]
    print(f"\nResiduals after iteration {i+1}:")
    print(df[["House Size", "House Price", "Prediction", "Residual"]])
    
    # Train a decision stump
    tree = DecisionTreeRegressor(max_depth=2, random_state=42)
    tree.fit(df[["House Size", "Num Bedrooms", "Year Built"]], df["Residual"])
    trees.append(tree)
    
    # Make predictions
    residual_predictions = tree.predict(df[["House Size", "Num Bedrooms", "Year Built"]])
    df["residual_predictions"] = residual_predictions
    df["Prediction"] += learning_rate * residual_predictions

    # Calculate MSE and store it
    mse_values.append(mean_squared_error(df["House Price"], df["Prediction"]))
    
    # Plot tree structure
    plt.figure(figsize=(10, 6))
    plt.title(f"Decision Tree {i+1}")
    plot_tree(tree, feature_names=["House Size", "Num Bedrooms", "Year Built"], filled=True)
    plt.show()
    
    print(f"\nUpdated Predictions after iteration {i+1}:")
    print(df[["House Size", "House Price", "Residual", "residual_predictions", "Prediction"]])

# Final comparison
plt.figure(figsize=(8, 6))
plt.scatter(df["House Size"], df["House Price"], color='blue', label='Actual Prices')
plt.scatter(df["House Size"], df["Prediction"], color='red', label='Final Predictions')
plt.xlabel("House Size")
plt.ylabel("House Price")
plt.title(f"Actual vs Final Predictions with learning rate {learning_rate} and {num_trees} trees built")
plt.legend()
plt.show()

# Plot MSE values over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(num_trees + 1), mse_values, marker='o', linestyle='-')
plt.xlabel("Number of Trees")
plt.ylabel("Mean Squared Error (MSE)")
plt.title(f"MSE vs. Number of Trees in Gradient Boosting with learning rate {learning_rate}")
plt.grid()
plt.show()


print("\nFinal Predictions vs Actual Prices:")
print(df[["House Size", "House Price", "Prediction"]])
```
